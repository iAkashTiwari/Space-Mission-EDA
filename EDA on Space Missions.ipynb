```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('Space Missions.xlsx')
```

```python
# Number of space mission by organizations
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Organisation')
plt.title('Number of Space Missions by Organisation')
plt.xticks(rotation=90)
plt.show()
```

```python
# Pie chart for rocket status
plt.figure(figsize=(6, 6))
df['Rocket_Status'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Percentage of Rocket Status')
plt.show()
```

```python
# Mission status by orginization
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='Organisation', hue='Mission_Status')
plt.title('Mission Status by Organisation')
plt.xticks(rotation=90)
plt.legend(title='Mission Status', loc='upper right')
plt.show()
```

```python
# Mission status
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='Mission_Status')
plt.title('Mission Status')
plt.show()
```

```python
# Box Plot of Mission Years by Organisation
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='Organisation', y='Year')
plt.title('Mission Years by Organisation')
plt.xticks(rotation=90)
plt.show()
```

```python
# Line chart of Number of Space Missions Over the Years
plt.figure(figsize=(10, 6))
missions_by_year = df['Year'].value_counts().sort_index()
missions_by_year.plot(kind='line', marker='o')
plt.title('Number of Space Missions Over the Years')
plt.xlabel('Year')
plt.ylabel('Number of Missions')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Rocket_Status', hue='Mission_Status')
plt.title('Mission Status Distribution for each Rocket Status')
plt.legend(title='Mission Status', loc='upper right')
plt.show()
```
